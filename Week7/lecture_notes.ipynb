* <span style="color:red">Normalization</span>
<br>

    * Verileri bir veritabanında düzenleme işlemi Normalleştirme olarak bilinir. Değerlerin kaydırıldığı ve 0 ile 1 aralığında ölçeklendiği fazlalığı azaltmak için kullanılan bir ölçekleme tekniğidir

* <span style="color:red">Standardation</span>
<br> 

    * Veri standardizasyonu, verilerin tek tip bir formatta yeniden yapılandırıldığı bir süreçtir. İstatistikte standardizasyon, tüm değişkenleri aynı ölçeğe koyarak değişkenleri karşılaştırır. Özelliklerin ortalamadan çıkarılıp standart sapmaya bölünmesiyle dönüştürülmesiyle yapılır. Bu süreç aynı zamanda Z-skoru olarak da bilinir.

* <span style="color:red">Batch Normalization</span>
<br>

* Batch normalization sayesinde ağdaki katmanlar, önceki katmanın öğrenmesini beklemek zorunda kalmaz. Eş zamanlı olarak öğrenime olanak sağlar. Eğitimimizin hızlanmasını sağlar.
* Ancak batch norm ile bir katmandaki değişiklik diğer katmana yayılmadığı için daha yüksek learning rate’ler kullanabiliriz. Ayrıca batch norm, ağın daha kararlı ve düzenli hale gelmesini sağlar.
