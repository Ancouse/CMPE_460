## Activation Functions

*  Sigmoid or Logistic Activation Function
    * Sigmoid Fonksiyon eğrisi bir S şekline benziyor
    * Sigmoid fonksiyonunu kullanmamızın ana sebebi (0 ile 1) arasında bulunmasıdır.
    * Fonksiyon monotondur ancak fonksiyonun türevi değildir.

<br>

* Tanh veya hiperbolik tanjant Aktivasyon Fonksiyonu
    * tanh ayrıca lojistik sigmoid gibidir ama daha iyidir. Tanh fonksiyonunun aralığı (-1 ila 1) arasındadır. tanh ayrıca sigmoidaldir (s - şeklindedir).
    * Bunun avantajı, negatif girdilerin güçlü bir şekilde negatif olarak eşlenmesi ve sıfır girdilerin tanh grafiğinde sıfıra yakın eşlenmesidir.
    * Fonksiyon türevlenebilir
    * Fonksiyon monoton iken türevi monoton değildir .
    * Tanh fonksiyonu esas olarak iki sınıf arasında sınıflandırma yapmak için kullanılır.

<br>

* ReLU (Rektifiye Doğrusal Birim) Aktivasyon Fonksiyonu
    * ReLU şu anda dünyada en çok kullanılan aktivasyon fonksiyonudur. Çünkü neredeyse tüm evrişimli sinir ağlarında veya derin öğrenmede kullanılmaktadır.
    * ReLU yarı düzeltilmiştir (alttan). z sıfırdan küçük olduğunda f(z) sıfırdır ve z sıfırın üzerinde veya sıfıra eşit olduğunda f(z) z'ye eşittir.
    * Aralık: [ 0 ila sonsuza kadar)
    * Fonksiyon ve türevinin her ikisi de monotondur.
    * Asıl sorun -> Tüm negatifleri 0 kabul etmesidir.

<br>

* Leaky ReLU
    * Ölmekte olan ReLU problemini çözme girişimidir.
    * Sızıntı, ReLU işlevinin aralığını artırmaya yardımcı olur. Genellikle, a'nın değeri 0,01 kadardır. a-> negatiflerin çarpanıdır. Yani artık onlar 0 değildir.
    * Bu nedenle, Sızdıran ReLU'nun aralığı (-sonsuzdan sonsuza) şeklindedir.
    * Hem Sızdıran hem de Rastgeleleştirilmiş ReLU işlevleri, doğası gereği monotondur. Ayrıca türevleri de monoton yapıdadır.
