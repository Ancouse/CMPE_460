What is a gradient ?

* Belirli bir yönde belirli bir noktada bir eğrinin eğimidir.
* Tek değişkenli bir fonksiyon durumunda, sadece seçilen bir noktadaki ilk türevdir.

* Çok değişkenli bir fonksiyon durumunda, her ana yönde (değişken eksenler boyunca) bir türev vektörüdür. Sadece bir eksen boyunca bir eğimle ilgilendiğimiz ve diğerlerini umursamadığımız için bu türevlere kısmi türevler denir.
